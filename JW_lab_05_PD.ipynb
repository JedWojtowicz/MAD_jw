```python
import pandas as pd
import os
```

```python
url = 'https://raw.githubusercontent.com/kropiak/swps-metody-analizy-danych/refs/heads/main/lab_05/data/loan_data.csv'
df = pd.read_csv(url)
df.index = range(1, df.shape[0] + 1)
```

## Zadania

1. Zamień nazwę kolumny `loan_amnt` na `loan_amount`.

```python
df.rename(columns={"loan_amnt": "loan_amount"},inplace=True)
df
```

2. Z ramki danych wykorzystanej w przykładach (loan) wyświetl:
* a) tylko kolumny 'person_gender', 'person_age' oraz 'person_income',
* b) wiersze od 1000 do 2000 włącznie (chodzi o kolejne wiersze w ramce, a nie wartość indeksu),
* c) wiersze parzyste,
* d) wiersze o parzystej wartości w kolumnie indeks,
* e) pierwszą i ostatnią kolumnę.

```python
print("a)")
display(df.loc[:,["person_gender", "person_age", "person_income"]])
print("b)")
display(df.loc[1000:2000])
print("c)")
display(df.loc[::2])
print("d)")
display(df.loc[df.index % 2 == 0])
print("e)")
display(df.iloc[:,[0,-1]])
```

3. Z tej samej ramki danych wyświetl dane wykorzystując filtry oparte o mapowanie:
* a) tylko kobiety z wykształceniem 'Master' oraz 'Bachelor' i pożyczkami pomiędzy 50000 a 70000,
* b) wszyscy mężczyźni z wykształceniem innym niż 'Master' oraz będący właścicielem lokalu (kolumna person_home_ownership = OWN),
* c) wszystkie osoby z wartością `loan_percent_income` w przedziale od 0.4 do 0.6,
* d) wszystkie wartości pożyczek, których kwota jest wielokrotnością liczby 10000.

```python
print("a)")
display(df[(df['loan_amount'] > 50000) & (df['loan_amount'] < 70000) & (df['person_gender'] == 'female') & (df['person_education'].isin(["Master", "Bachelor"]))])
print("b)")
display(df[(df['person_gender'] == 'male') & (df['person_home_ownership'] == 'OWN') & ~(df['person_education'].isin(["Master"]))])
print("c)")
display(df[(df['loan_percent_income'] >= 0.40) & (df['loan_percent_income'] <= 0.60)])
print("d)")
display(df[df['loan_amount'] % 10000 == 0])
```
