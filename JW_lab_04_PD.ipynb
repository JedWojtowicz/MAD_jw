Stwórz z tych danych ramkę pandas. Kolumny nazwij wg. schematu ['base', 'pow_2', 'pow_3', ..., 'pow_10']

````python
names = ['base']
for i in range(2, 11):
    names.append(f'pow_{i}')

arr = np.arange(1, 11).reshape(10, 1)
arr



arr = np.hstack([arr ** i for i in range(1, 11)])
arr

df = pd.DataFrame(arr, columns=names)
df
````
